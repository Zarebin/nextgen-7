```python
test = df.at[0,'html']
```

```python
def decode_base64(encoded_html, encoding = 'utf-8'):
    base64_bytes = encoded_html.encode(encoding)
    message_bytes = base64.b64decode(base64_bytes)
    decoded = message_bytes.decode(encoding)
    return decoded

decoded = decode_base64(test)
decoded
```

```python
parsed = BeautifulSoup(decoded, "html.parser")
print(parsed.prettify())
```

```python
def find_og_description(parsed_html):
   tag = parsed_html.find("meta", property="og:description")
   return tag['content']
```
